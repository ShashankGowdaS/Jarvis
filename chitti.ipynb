```python
class MainThread(QThread):
    def __init__(self):
        super(MainThread,self).__init__()
        
    def run(self):
        self.TaskExecution()

        
    def takeCommand(self):
      
    # it takes microphone input from users and returns string output

        r=sr.Recognizer()

        with sr.Microphone() as source:
            print('Listening.........')
            r.pause_threshold=1
            audio = r.listen(source)

        try:
            print('Recognizing.....')
            query = r.recognize_google(audio) 
            #print(f"User said:{query}\n")
            print('USER said:',query)

        except Exception as e: 
            print(e)
            print('Say that again please....')
            return "None"

        return query
    
    def speak(self,audio):
        engine.say(audio)
        engine.runAndWait()

    def wishMe(self):
        hour= int(datetime.datetime.now().hour)
        if hour>=0 and hour<12:
            self.speak("Good morning")

        elif hour>=12 and hour<18:
            self.speak("Good Afternoon")

        else:
            self.speak("Good Evening!")

        #self.speak("I am Chhittti the Robot       Speed 1 tera byte...Memory 1 zetaa byte")    
        self.speak("I am Jarvis..Tell me how may I help You")    
    
    def TaskExecution(self):
       
        #if __name__=="__main__":
        self.wishMe()

        sht=True
        while sht:
        #if 1:
            self.query=self.takeCommand().lower()

        #logic for executing tasks on self.query

            if 'wikipedia' in self.query:
                self.speak('Searching wikipedia...')
                print('it takes a little while')
                try:
                    self.query=self.query.replace('wikipedia','')
                    results=wikipedia.summary(self.query,sentences=2)
                    print("According to wikipedia")
                    self.speak("According to wikipedia")
                    print(results)
                    self.speak(results)
                except Exception as e:
                    print(e)
                    self.speak('Would you mind trying it again Sir???')


            elif   'open youtube' in self.query:
                print('opening youtube sir')
                webbrowser.open('youtube.com')


            elif 'open google' in self.query:
                webbrowser.open('google.com')



            elif 'play music' in self.query:
                music_dir="D:/MultiMedia/Music"
                songs=os.listdir(music_dir)
                #print(songs)
                os.startfile(os.path.join(music_dir,songs[15]))

            elif 'the time' in self.query:
                strTime=datetime.datetime.now().strftime("%H:%M:%S")
                print(f"Sir, the time is {strTime}")
                self.speak(f"Sir, the time is {strTime}")



            elif 'open code' in self.query:
                codePath="C:/Users/Shashank.S/AppData/Local/Programs/Microsoft VS Code/Code.exe"
                os.startfile(codePath)


            elif 'email' in self.query:
                try:
                    speak('Whom should I send it to , Sir??')
                    toName=takeCommand()
                    speak('What should I say?')
                    content=takeCommand()
                    to='toName'+'@gmail.com'
                    sendEmail(to,content)
                    speak('Email has been sent')

                except Exception as e:    
                    print(e)
                    speak('Sorry buddy, I couldnt send this email')

            elif 'shut down'or'shutdown' in self.query:
                self.speak('Shutting down.....Jarvis going to sleep')
                print('Shutting down')
                #break
                sht=False     
            else: 
                continue 


        
        
        
        
startExecution=MainThread()        
              
        
class Main(QMainWindow):
    def __init__(self):
        super().__init__()
        self.ui=Ui_JARVIS()
        self.ui.setupUi(self)
        self.ui.pushButton.clicked.connect(self.startTask)
        self.ui.pushButton_2.clicked.connect(self.close)
        
        
        
    def startTask(self):
        self.ui.movie=QtGui.QMovie("jarvis gif.gif")
        self.ui.label.setMovie(self.ui.movie)
        self.ui.movie.start()
        self.ui.movie_2=QtGui.QMovie("initiating jarvis.gif")
        self.ui.label.setMovie(self.ui.movie_2)
        self.ui.movie.start()
        timer=QTimer(self)
        timer.timeout.connect(self.showTime)
        timer.start()
        
        startExecution.start()
        
    def showTime(self):
        curtime=QTime.currentTime()
        curdate=QDate.currentDate()
        labeltime=curtime.toString('hh:mm:ss')
        labeldate=curdate.toString(Qt.ISODate)

        self.ui.textBrowser.setText(labeldate)
        self.ui.textBrowser_2.setText(labeltime)
            
            
            
app=QApplication(sys.argv)           
jarvis=Main()
jarvis.show()
exit(app.exec_())
```
